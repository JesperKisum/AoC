```python
from aocd import get_data, submit
from aocd.models import Puzzle
import os
from datetime import datetime
with open('../secret.key', 'r') as f:
    os.environ['AOC_SESSION'] = f.read().strip()

def parse_data(day):
    input = get_data(day=day, year=2024)
    
    data = [[int(x) for x in line] for line in input.split('\n')]
    return data
day = datetime.now().day
data = parse_data(day)
puzzle = Puzzle(year=2024, day=day)

#puzzle.view()
print(puzzle.title)
print(data)
```

```python
def part1():

    def findscore(i,j):
        
        visted=set()
        score_sum=[0]
        def dfs(i,j,prev):
            
            if (i,j) in visted:
                return
            
            if not (0 <= i < len(data) and 0 <= j < len(data[0])):
                return 
            if not ((prev+1) == data[i][j]):
                return
            
            visted.add((i,j))

            if data[i][j] == 9:
                score_sum[0]+=1
                return
            
            dfs(i + 1, j, data[i][j])
            dfs(i - 1, j, data[i][j])
            dfs(i, j + 1, data[i][j])
            dfs(i, j - 1, data[i][j])

        dfs(i + 1, j, data[i][j])
        dfs(i - 1, j, data[i][j])
        dfs(i, j + 1, data[i][j])
        dfs(i, j - 1, data[i][j])
        return score_sum[0]
    
    count=0
    for i in range(len(data)):
        for j in range(len(data[0])):
            if data[i][j]==0:
                count+=findscore(i,j)
            
    return count
part1()

```
