```python
from collections import deque 


def get_disk(data):
    disk = deque()
    id = 0
    for idx, val in enumerate(data):
        val = int(val)
        if idx % 2 == 0:
            for i in range(val):
                disk.append(id)
            id += 1
        else:
            for i in range(val):
                disk.append(".")

    return disk, id

def find_and_replace(rev_disk):
    if len(rev_disk) == 0:
            return rev_disk, None, None 
    
    while rev_disk[-1] == ".":
        rev_disk.pop()
        
    value = rev_disk[-1]
    rev_disk.pop()
    return rev_disk, value, len(rev_disk)
    

def fragment_disk(disk, rev_disk):
    dot_count = disk.count(".")
    num_count = len(disk) - dot_count
    count = 0
    val = None
    for idx, i in enumerate(disk):
        if idx+1 == num_count:
            break
        if isinstance(i, str):
            rev_disk, val, remove = find_and_replace(rev_disk)
            disk[idx] = val    
            disk[remove] = "."
            count += 1
            if count == dot_count:
                break
                    
    return disk

def calc_checksum(disk):
    checksum = 0
    for idx , i in enumerate(disk):
        if isinstance(i, int):
            checksum += idx * i
    return checksum

disk, _ = get_disk(test)
rev_disk = disk.copy()
disk = fragment_disk(disk, rev_disk)


print(calc_checksum(disk))
#puzzle.answer_a = calc_checksum(disk)
print(disk)
```

```python
def find_file(rev_disk, file_id):
    if len(rev_disk) == 0:
            return rev_disk, None, None
    while rev_disk[-1] == ".":
        rev_disk.pop()
    file_size = 0    
    while rev_disk[-1] == file_id-1:
        rev_disk.pop()
        file_size += 1
    return rev_disk, file_id-1, file_size

def find_free_slot(disk, file_size):

    for idx, i in enumerate(disk):
        if isinstance(i, str):
            if list(disk)[idx:idx+file_size] == ["."]*(file_size):
                return idx
    return None


disk, max_id = get_disk(data)
rev_disk = disk.copy()


files = []
print("starting to find files")
for id in range(0, max_id):
    file_id = max_id - id
    try:
        rev_disk, file_id, size = find_file(rev_disk, file_id)
    except IndexError:
        break
    if size:
        files.append([file_id, size])
print("files found", files)

print("starting to sort files")
for file in files:
    file_id, size = file 
    slot = find_free_slot(disk, size)
    if file_id % 100 == 0:
        print(file_id/100)
    if slot and slot > disk.index(file_id):
        slot = None
    if slot is not None:
        while file_id in disk:
            disk[disk.index(file_id)] = "."
        for i in range(size):
            disk[slot+i] = file_id

print(disk)
calc_checksum(disk)
```
