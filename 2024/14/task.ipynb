```python
W = 101
H = 103

ans = [0,0,0,0]

st = 0
for line in data:
    if line.strip() == "":
        continue
    p,v = line.split()
    px,py = map(int,p[2:].split(",")) 
    vx,vy = map(int,v[2:].split(","))
    nx,ny = px + 100*vx, py + 100*vy
    nx = nx % W
    ny = ny % H
    if nx == W//2 or ny == H//2:
        continue
    if nx < W//2 and ny < H//2:
        ans[0] += 1
    elif nx > W//2 and ny < H//2:
        ans[1] += 1
    elif nx < W//2 and ny > H//2:
        ans[2] += 1
    else:
        ans[3] += 1
        
print(ans)
print(ans[0] * ans[1] * ans[2] * ans[3])
```

```python

import time

W = 101
H = 103

ans = [0,0,0,0]

robots = []

st = 0
for line in data:
    if line.strip() == "":
        continue
    p,v = line.split()
    px,py = map(int,p[2:].split(",")) 
    vx,vy = map(int,v[2:].split(","))
    robots.append(((px,py),(vx,vy)))

seconds = 0
while True:
    grid = [[0 for _ in range(W)] for _ in range(H)]
    seconds += 1

    bad = False
    for robot in robots:
        pr1,pr2 = robot
        px,py = pr1
        vx,vy = pr2
        nx,ny = px + seconds*vx, py + seconds*vy
        nx = nx % W
        ny = ny % H
        grid[ny][nx] += 1
        if grid[ny][nx] > 1:
            bad = True

    if not bad:
        print(seconds)
        for row in grid:
            print("".join(map(str,row)))
        break
```
